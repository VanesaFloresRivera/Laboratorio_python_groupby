```python
df_netflix_agrupado = pd.read_csv("datos/Netflix_full.csv", index_col=0)
df_netflix_agrupado.head(5)
```

#### Ejercicio 1: ¿Cuántos títulos hay por cada tipo de contenido ('Movie' o 'TV Show')?

```python
df_recuento_tipo_contenido = df_netflix_agrupado.groupby("type")["title"].count()
df_recuento_tipo_contenido = df_recuento_tipo_contenido.reset_index()
df_recuento_tipo_contenido
```

```python
df_recuento_tipo_contenido['%'] = np.round(df_recuento_tipo_contenido.title/df_netflix_agrupado.shape[0]*100,2)
df_recuento_tipo_contenido
```

#### Ejercicio 2: ¿Cuál es el promedio de puntuación de las películas agrupadas por clasificación de audiencia ('rating')? Interpreta los resultados, ¿hay algo que te llame la atención? Explica por qué. 

```python
df_rating_promediopuntuacion = df_netflix_agrupado.groupby('rating')["IMDB Score"].mean().reset_index()
df_rating_promediopuntuacion
```

Todas tienen un promedio entre 6 y 7. No obstante, las categorioas NC-17,NR, TV-Y7-FV y UR no tienen información porque solo tenemos puntuación para las películas originales de Netflix y en las originales estas categorías no aparecían.

#### Ejercicio 3: ¿Cuántos títulos hay por cada país? Ordena los resultados de mayor a menor

```python
df_contador_paises = df_netflix_agrupado.groupby('country')['title'].count().reset_index().sort_values('title',ascending=False)
df_contador_paises

df_contador_paises['%'] = (df_contador_paises.title/df_netflix_agrupado.shape[0]*100).round(2)
df_contador_paises
```

#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
df_netflix_agrupado.date_added = pd.to_datetime(df_netflix_agrupado.date_added) # para convertir la columna en fecha
#display(df_netflix_agrupado.info())
df_netflix_agrupado["año"] = df_netflix_agrupado['date_added'].dt.year #para quedarme solo con el año
#display(df_netflix_agrupado.info())

df_contador_año = df_netflix_agrupado.groupby('año')['title'].count().reset_index()
df_contador_año['%'] = (df_contador_año.title/df_netflix_agrupado.shape[0]*100).round(2)
df_contador_año
```

#### Ejercicio 8: ¿Cuántos títulos fueron lanzados en cada año agrupados por tipo ('Movie' o 'TV Show')?

```python
df_añolanzamiento_tipo= df_netflix_agrupado.groupby(['release_year', 'type'])['title'].count().reset_index()
df_añolanzamiento_tipo
```

```python
#con unstack
serie_añolanzamiento_type = df_netflix_agrupado.groupby(['release_year', 'type'])['title'].count()
df_añolanzamiento_tipo_1= serie_añolanzamiento_type.unstack()
df_añolanzamiento_tipo_1
```

```python
df_añolanzamiento_tipo_1['%Movie'] =np.round( df_añolanzamiento_tipo_1['Movie']/(df_añolanzamiento_tipo_1['Movie']+df_añolanzamiento_tipo_1['TV Show'])*100,2)
df_añolanzamiento_tipo_1['%TV Show'] = np.round( df_añolanzamiento_tipo_1['TV Show']/(df_añolanzamiento_tipo_1['Movie']+df_añolanzamiento_tipo_1['TV Show'])*100,2)
df_añolanzamiento_tipo_1
```

#### Ejercicio 9: ¿Cuál es el país con la mayor cantidad de títulos de clasificación 'TV-MA'?
